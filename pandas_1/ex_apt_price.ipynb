```python
# melt를 이용하여 월별데이터를 '기간' 컬럼으로 가격은 '평당분양가격'으로 처리하여 first_df_melt로 저장
first_df_melt = first_df.melt(id_vars='지역',var_name='기간',value_name='평당분양가격')
first_df_melt.head(3)
```

```python
# 기간 컬럼을 연도,월 컬럼으로 분리
first_df_melt['연도'] = first_df_melt['기간'].str.split('년').str[0].astype(int)
first_df_melt['월'] = first_df_melt['기간'].str.split('년').str[1].str.replace('월','').astype(int)
```

```python
first_df_melt.head(3)
```

```python
first_df_melt.info()
```

```python
# last_df 전용면적에서 전체 인 데이터만 last_df_prepare로 저장
last_df_prepare = last_df[last_df['전용면적']=='전체'].copy()
```

```python
last_df_prepare.head(2)
```

```python
first_df_melt.head(2)
```

```python
# 지역명-> 지역 컬럼으로 변경
```

```python
last_df_prepare.columns =['지역명', '연도', '월', '분양가격', '평당분양가격', '전용면적']
```

```python
last_df_prepare.rename(columns={'지역명':'지역'},inplace=True)
```

```python
last_df_prepare.head(2)
```

```python
#concat으로 데이터 합치기
first = first_df_melt[['지역','연도','월','평당분양가격']]
last = last_df_prepare[['지역','연도','월','평당분양가격']]
df = pd.concat([first,last])
df.reset_index(drop=True,inplace=True)
df
```

```python
df.groupby(['연도','월'])['평당분양가격'].mean()
```
